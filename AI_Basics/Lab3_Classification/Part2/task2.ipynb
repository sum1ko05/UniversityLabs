Getting Iris data, splitting it and training a model

```python
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression

iris = load_iris()
x = iris.data[:, 2:]
y = iris.target

clf = LogisticRegression(random_state=0)
# No auto tests needed
clf.fit(x, y)
```

Displaying our results on plot

```python
import numpy as np
import matplotlib.pyplot as plt

x1_min, x1_max = x[:, 0].min() - 1, x[:, 0].max() + 1
x2_min, x2_max = x[:, 1].min() - 1, x[:, 1].max() + 1

xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, 0.01), np.arange(x2_min, x2_max, 0.01))
yy = clf.predict(np.c_[xx1.ravel(), xx2.ravel()])
yy = yy.reshape(xx1.shape)

fig, ax = plt.subplots()

ax.set_title('Multiclass logistic regression')
ax.set_xlabel('Petal length (cm)')
ax.set_ylabel('Sepal length (cm)')

ax.contourf(xx1, xx2, yy, alpha=0.3, cmap='viridis')
ax.scatter(
            x=x[:, 0], y=x[:, 1],
            c=y, cmap='viridis', edgecolor='k'
          )

plt.show()
```
